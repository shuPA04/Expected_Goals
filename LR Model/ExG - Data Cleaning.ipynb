```python
import pandas as pd
import glob
pd.set_option('display.max_columns', None)
```

```python
#extract file names
path = "Statsbomb/data/matches/"

competition_id = 11

path_to_json = path + str(competition_id)

file_list = glob.glob(path_to_json + "/*.json")

print(file_list)
```

```python
#create pandas dataframes from the json files
dfs = []

for file in file_list:
    data = pd.read_json(file)
    dfs.append(data)
    

combined = pd.concat(dfs, ignore_index=True)
```

```python
combined['match_id'].count()
```

```python
match_ids = combined["match_id"].to_list()
```

```python
len(match_ids)
```

```python
#extract event data according to match_id
path_to_events_json = "Statsbomb/data/events/"

event_lists = []

_list = glob.glob(path_to_events_json + "*.json")

for x in _list:
    for i in range(len(match_ids)):
        if x == path_to_events_json + str(match_ids[i]) + ".json":
            event_lists.append(x)
```

```python
print(event_lists)
print(len(event_lists))
```

```python
#create pandas dataframes from the json files containing event data
event_dfs = []

for file in event_lists:
    event_data = pd.read_json(file)
    event_dfs.append(event_data)
    
event_df = pd.concat(event_dfs, ignore_index=True)
```

```python
event_df.head()
```

```python
df_shot = event_df.dropna(axis=0, subset=['shot'])

df_shot.head()
```

```python
df_location = df_shot['location']
```

```python
df_shot_modified = pd.DataFrame(df_shot['shot'].tolist())
```

```python
df_shot_modified.head()
```

```python
df_dict = df_shot_modified[['technique', 'body_part', 'type', 'outcome']]
```

```python
shot_data = df_shot_modified.drop(['technique', 'body_part', 'type', 'outcome', 'end_location', 'freeze_frame', 'key_pass_id'], axis=1)
```

```python
df_dict_technique = pd.DataFrame(df_dict['technique'].tolist())
```

```python
df_dict_technique.head()
```

```python
df_technique = df_dict_technique.drop('id', axis=1)
df_technique.rename(columns={"name":"technique"}, inplace=True)
print(df_technique.head())
print(len(df_technique))
```

```python
df_dict_body_part = pd.DataFrame(df_dict['body_part'].tolist())
```

```python
df_dict_body_part.head()
```

```python
df_body_part = df_dict_body_part.drop('id', axis=1)
df_body_part.rename(columns={"name":"body_part"}, inplace=True)
print(df_body_part.head())
print(len(df_body_part))
```

```python
df_dict_type = pd.DataFrame(df_dict['type'].tolist())
```

```python
df_dict_type.head()
```

```python
df_type = df_dict_type.drop('id', axis=1)
df_type.rename(columns={"name":"type"}, inplace=True)
print(df_type.head())
print(len(df_type))
```

```python
df_dict_outcome = pd.DataFrame(df_dict['outcome'].tolist())
```

```python
df_dict_outcome.head()
```

```python
df_outcome = df_dict_outcome.drop('id', axis=1)
df_outcome.rename(columns={"name":"outcome"}, inplace=True)
print(df_outcome.head())
print(len(df_outcome))
```

```python
df_shot_final = pd.concat([shot_data, df_technique, df_body_part, df_type, df_outcome, df_location.reset_index(drop=True)], axis=1)
```

```python
df_shot_final.head()
```
